```python
# Python Program to Add Two Matrices

# Function for matrix addition
def add_matrix(mat1, mat2):

  rows, cols = (len(mat1),len(mat1[0]))
  mat3=[[0 for i in range(cols)] for j in range(rows)] # initialize the result matrix


  for i in range(len(mat1)):
    for j in range(len(mat1[i])):
            mat3[i][j]= mat1[i][j] + mat2[i][j]

  return mat3



# Function for displaying the list as a matrix to the user
def display_matrix(mat):
  for i in range(len(mat)):
    print(mat[i])
  

# x and y are the matrices to add
x=[[1,2,3,8],
   [2,3,4,7],
   [5,6,7,5]]


y=[[4,5,8,5],
   [6,7,9,0],
   [4,5,7,3]]

# This code assumes the matrices are of same size

z=add_matrix(x, y)

# display the result
display_matrix(x)
print("\n+\n")
display_matrix(y)
print("\n=\n")
display_matrix(z)
```

```python
# Python Program to Multiply Two Matrices
# This program works with 3 x 3 matrix

# Function for matrix multiplication
def multiply_matrix(mat1, mat2):
  
  rows, cols = (len(mat1), len(mat2[0]))
  
  mat3=[[0 for i in range(cols)] for j in range(rows)] # initialize the result matrix

  for i in range(len(mat1)): 
    for j in range(len(mat2[0])):
        for k in range(len(mat2)):
            mat3[i][j]= mat3[i][j]+(mat1[i][k]*mat2[k][j])
  return mat3





# Function for displaying the list as a matrix to the user
def display_matrix(mat):
  for i in range(len(mat)):
    print(mat[i])
  

# x and y are the matrices to multiply
x=[[1,2,3,5],
   [2,3,4,6],
   [5,6,7,7]]


y=[[4,5,8],
   [6,7,9],
   [4,5,7],
   [2,3,4]]

if len(x[0])==len(y): # check matrices are compatible for multiplication
  # This code assumes the elements of list are of equal size

  z=multiply_matrix(x, y)

  # display the result
  display_matrix(x)
  print("\nx\n")
  display_matrix(y)
  print("\n=\n")
  display_matrix(z)
else:
  print("no.of col of first matrix must be equal to  no.of rows of second matrix \n")
```

```python
#Python Program to Transpose a Matrix

# Function for matrix transpose
def transpose_matrix(mat):
  
  rows, cols = (len(mat[0]), len(mat))
  mat_trans=[[0 for i in range(cols)] for j in range(rows)] # initialize the result matrix

  for i in range(len(mat)):# transpose the matrix
    for j in range(len(mat[i])):
       mat_trans[j][i]= mat[i][j]
  return mat_trans


# Function for displaying the list as a matrix to the user
def display_matrix(mat):
  for i in range(len(mat)):
    print(mat[i])
  

# x is the matrix to transpose
x=[[1,2,3,4],
   [2,3,4,8],
   [5,6,7,7]]

y=transpose_matrix(x)

# display the result
display_matrix(x)
print("\nTransposed Matrix\n")
display_matrix(y)
```

```python
# Python Program to Sort Words in Alphabetic Order
input_string=input("Enter the list of words to sort. Press enter when you finish : ")
input_string=input_string.lower()  # convert all words to lower case

words = input_string.split()  
words.sort()
print("The sorted lis is : ")
for word in words:
  print (word.capitalize())
  
```
