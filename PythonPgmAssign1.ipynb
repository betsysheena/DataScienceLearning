```python
# 2. Python program to do arithmetical operations addition and division.
print("Enter two numbers for addition")
num1=int(input("Enter First number:"))
num2=int(input("Enter Second number "))
sum=num1 + num2
print(f"The sum is", sum)

print("Enter two numbers for division")
num3=int(input("Enter First number:"))
num4=int(input("Enter Second number "))

#Force the user to enter a non zero number
while num4 == 0:
  print("Division with zero is not allowed")
  num4=int(input("Enter another number "))

div_ans=num3 / num4
print(f"The answer is {div_ans}")
```

```python
# Python program to find the area of a triangle
height=int(input("Enter the height of the triangle "))
base=int(input("Ënther the length of the base "))
tri_area=0.5 * height * base
print(f"The area of the triangle is {tri_area}" )
```

```python
# Python program to swap two variables
var1=input("Input variable 1 : ")
var2=input("Input variable 2 : ")
var3=var1
var1=var2
var2=var3
print("Variables Swapped")
print(f"Variable 1 = {var1}" )
print(f"Variable 2 ={var2}")
```
