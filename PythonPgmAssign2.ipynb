```python
# Python program to convert kilometers to miles
kilomtr=float(input("Enter distance in kilometers "))
# 1 kilometer = 0.62137119 miles.
miles=round( kilomtr * 0.62137119 ,2)
print(f"Distance in miles : {miles} miles")
```

```python
#  Python program to convert Celsius to Fahrenheit
celsius=float(input("Enter temperature in Celsius "))
# To convert Celsius to Fahrenheit, multiply by 1.8 (or 9/5) and add 32.
fahrenheit=round( celsius * 1.8 + 32, 2)
print(f"Temperature in Fahrenheit : {fahrenheit} F")
```

```python
#  Python program to display calendar
import calendar

yy = int(input("Enter year: "))
mm = int(input("Enter month: "))

# display the calendar
print(calendar.month(yy, mm))
```
