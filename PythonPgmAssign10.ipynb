```python
#  Python program to find sum of elements in list
def add_list(mylist):
  sum=0
  for i in len(mylist):
    sum=sum+i
  return sum

my_list=[23,45,56,78,89,90]
print(f"Sum of list {my_list} using in-built function is {sum(my_list)}")
print(f"Sum of list {my_list} using  function is {sum(my_list)}")
```

```python
# Python program to Multiply all numbers in the list
def multiply_list(mylist):
  product=1
  for i in mylist:
    product=product*i
  return product

my_list=[1,2,5,4]
print(f"Product of elements  of list {my_list}  is {multiply_list(my_list)}")
```

```python
# Python program to find smallest number in a list

def min_list(mylist):
  min=mylist[0]
  for i in mylist:
    if i<min:
     min=i
  return min


my_list=[23,45,56,78,89,9]
print(f"Smallest element of list {my_list} using in-built function is {min(my_list)}")
print(f"Smallest element of list {my_list} using  function is {min_list(my_list)}")
```

```python
# Python program to find largest number in a list
def max_list(mylist):
  max=mylist[0]
  for i in mylist:
    if i>max:
      max=i
  return max


my_list=[23,45,56,78,89,9]
print(f"Largest element of list {my_list} using in-built function is {max(my_list)}")
print(f"Largest element of list {my_list} using  function is {max_list(my_list)}")
```

```python
# Python program to find second largest number in a list
def second_max_list(mylist):
  max=mylist[0]
  last_index=0
  index=0
  for i in range(0,len(mylist)):
    if mylist[i]>max:
      max=mylist[i]
      last_index=index
      index=i
  return mylist[last_index]


my_list=[23,45,56,78,89,91]
print(f"Second largest element of list {my_list} is {second_max_list(my_list)}")
```

```python
#  Python program to find N largest elements from a list
def n_largest_list(mylist,n):
  mylist.sort(reverse=True)
  return(mylist[n-1])

my_list=[72,45,78,6,4,7]
n=3

if n<=len(my_list):
  print(f"{n} largest element of list {my_list} is {n_largest_list(my_list,n)}")
else:
  print("Position should be less than the length of the list")
```

```python
#Python program to print even numbers in a list
my_list =[72,45,78,6,4,7,23,45,56,73,89,90]
even =[i for i in my_list if i%2==0]
print(f"Even numbers in the list {my_list} are {even}")
```

```python
# Python program to print odd numbers in a list
my_list =[72,45,78,6,4,7,23,45,56,73,89,90]
odd = [i for i in my_list if i%2!=0 ]
print(f"Odd numbers in the list {my_list} are {odd}")
```

```python
#Remove empty List from List
test_list = [75, 34, 23,[], 37, [],67, [], 98]
new_list=[i for i in test_list if i!=[]]
print(f"The original list is {test_list}")
print(f"The new list after removing empty list is {new_list}")
```

```python
# Python program to Cloning or Copying a list
first_list = [375, 348, 203, 397,6557, 978]
copy_list=[i for i in first_list]
print(f"The original list is {first_list}")
print(f"The copy is {copy_list}")
```
