```python
# Python Program to Find LCM
# LCM- Least Common Multiplier - will always be >= highest given number
# Starting from the highest given/input number (say x) start a loop, incrementing by 1 , till product of both numbers 
# Inside loop check x can be divided by both given numbers completly without reminders
# if it can, x is the LCM

num1=0
num2=0
# Accept positive numbers only as input
while num1<=0:
  num1=int(input("Enter first number(intiger >0) to Find LCM :" ))

while num2<=0:
  num2=int(input("Enter second  number(intiger >0)  to Find LCM :" ))

high= max(num1,num2)
prod=num1*num2
lcm=high

for x in range(high,prod+1):
  if(x%num1==0 and x%num2==0):
    lcm=x
    break;
print(f"LCM of {num1} and {num2} is {lcm}")


```

```python
#Python Program to Find HCF
# HCF-Highest Common Factor - Highest positive integer that perfectly divides the two given numbers
# HCF will always be <= smllest give/input number
# Starting from the lowest given/input number start a loop (say x), decrease  by 1 , till  reaches 1
# Inside loop check x can be divided by both given numbers completly without reminders
# if it can, x is the HCF

num1=0
num2=0
# Accept positive numbers only as input
while num1<=0:
  num1=int(input("Enter first number(intiger >0) to Find HCF :" ))

while num2<=0:
  num2=int(input("Enter second  number(intiger >0) to Find HCF :" ))

# HCF will always be <= smllest give/input number
small= min(num1,num2)

for x in range(small,0,-1):
  if(num1%x==0 and num2%x==0):
    hcf=x
    break;

print(f"HCF of {num1} and {num2} is {hcf}")

```

```python
#  Python Program to Convert Decimal to Binary, Octal and Hexadecimal?

# function for decimal to binary conversion 
def decimal_binary(decimal):
  deci=decimal #create a copy of the argument
  if deci==0:
    binary='0'
  else:
    binary='' # initialise an empty string
  while deci!=0:
    deci,remainder=deci//2, deci%2
    binary=binary+str(remainder)
    
  binary=binary[::-1] #reverse
  return binary


# function for decimal to octal conversion 
def decimal_octa(decimal):
   deci=decimal #create a copy of the argument
   if deci==0:
    octa='0'
   else:
    octa='' # initialise an empty string
   while deci!=0:
    deci,remainder=deci//8, deci%8
    octa=octa+str(remainder)
    
   octa=octa[::-1] #reverse
   return octa

# function for decimal to hexadecimal conversion 
def decimal_hexa(decimal):
   deci=decimal #create a copy of the argument
   conversion_table = { 10: 'A', 11: 'B', 12: 'C',  13: 'D', 14: 'E', 15: 'F'}

   if deci==0:
    hexa='0'
   else:
    hexa='' # initialise an empty string
   while deci!=0:
    deci,remainder=deci//16, deci%16
    if remainder<10:
      hexa=hexa+str(remainder)
    else:
      hexa=hexa+conversion_table[remainder]

   hexa=hexa[::-1] #reverse
   return hexa

##############################
num=-1
# Accept positive numbers only as input
while num<0:
  num=int(input("Enter a number :" ))
  
print(f"Binaary equvalent of {num} is {decimal_binary(num)}")
print(f"Octal equvalent of {num} is {decimal_octa(num)}")
print(f"Hexa equvalent of {num} is {decimal_hexa(num)}")
```
