```python
#  Python Program to find sum of array
# Method 1 : Using built in function
my_array=[1,2,3,4,5,6]
print(f"The array is {my_array}")
print(f"(Method 1)The sum of array is {sum(my_array)}")

## Using for loop
array_sum=0
for i in my_array:
  array_sum=array_sum+i
print(f"(Method 2)The sum of array is {array_sum}")
```

```python
#Python Program to find largest element in an array
#Method 1 : Using built in function
my_array=[1,2,3,4,5,6]
print(f"The array is {my_array}")
print(f"(Method 1)The largest element is {max(my_array)}")

## Using for loop
biggest=my_array[0]
for i in my_array:
  if i>biggest:
    biggest=i
print(f"(Method 2)The largest element is {biggest}")
```

```python
# Python Program for array rotation
#Method 1 : Using list reverse method
my_array=[1,2,3,4,5,6]
print(f"Original array {my_array}")
my_array.reverse()
print(f"The revered array {my_array}")

## Using slicing
my_array=[1,2,3,4,5,6]
print(f"Original array {my_array}")
rev_my_array=my_array[::-1]
print(f"The revered array {rev_my_array}")
```

```python
# Python Program to Split the array and add the first part to the end
my_array=[1,2,3,4,5,6]
print(f"Original array {my_array}")
position=4  # split at this position
part1=my_array[:position]
part2=my_array[position:]
my_array=part2+part1
print(f"The splitted at position {position} and rejoined array {my_array}")
```

```python
# Python Program to check if given array is Monotonic
# An array is monotonic  if it is either continuously increasing or continuously decreasing

my_array=[1,2,9,4,5,6] #not a monotone  array
# my_array=[1,2,3,4,5,6] #monotone increasing
# my_array=[34,23,11,7] #monotone decreasing 
# my_array=[7,7,7,7,7] # array with same element

# check whether the array is monotone increasing
monotonic_increase=True
same=False
i=0
length=len(my_array)-1

while i<length:
  if my_array[i]<=my_array[i+1]:
      monotonic_increase=True
  else:
      monotonic_increase=False
      break
  i=i+1
if monotonic_increase:
  if my_array[0]==my_array[length]: # check if all elements are same
    same=True
    monotonic_increase=False
  else:
    print(f"{my_array} is a monotone increasing array")

# check whether the array is monotone decreasing
i=0
monotonic_decrease=True

while i<length:
  if my_array[i]>=my_array[i+1]:
      monotonic_decrease=True
  else:
      monotonic_decrease=False
      break
  i=i+1
if monotonic_decrease:
   if my_array[0]==my_array[length]: # check if all elements are same
    same=True
    monotonic_decrease=False
   else:
    print(f"{my_array} is a monotone decreasing array")

if same:
  print(f"In {my_array} all elements are the same")

if not monotonic_increase and not monotonic_decrease:
  print(f"{my_array} is a not a monotone  array")

```
